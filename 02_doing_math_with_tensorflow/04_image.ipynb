```python
%matplotlib inline
import tensorflow as tf
import matplotlib.image as mp_image
import matplotlib.pyplot as plt
```

```python
filename = 'openvpn.png'
input_image = mp_image.imread(filename)
```

```python
print('dim:', input_image.ndim)
print('shape:', input_image.shape)
print('type:', input_image.dtype)
(height, width, colors_deep) = input_image.shape
```

```python
plt.imshow(input_image)
```

```python
my_image = tf.placeholder('float32', [None, None, 3])
# my_image = tf.Variable(input_image, name = 'img')
# Slice image from height 50 - 150px, full width, and full color depth (RGB)
img_slice = tf.slice(my_image, [50, 0, 0], [150, -1, -1])
```

```python
model = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(model)
    result = sess.run(img_slice, feed_dict = {my_image: input_image})
    print(result.shape)
```

```python
plt.imshow(result)
```

```python
# Transpose image

x = tf.Variable(input_image, name = 'x')
model = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(model)
    x = tf.transpose(x, perm = [1, 0, 2])
    result = sess.run(x)
plt.imshow(result)
```

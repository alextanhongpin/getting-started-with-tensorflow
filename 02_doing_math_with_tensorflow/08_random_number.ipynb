```python
# %matplotlib --list
%matplotlib notebook
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
uniform = tf.random_uniform([100], # The number of elements
                            minval = 0, # The min range
                            maxval = 1, # The max range
                            dtype = tf.float32) # The type
```

```python
with tf.Session() as sess:
    print(uniform.eval())
    plt.hist(uniform.eval(), normed = True)
```
